Finally, we have reached the point where we can discuss Mixtures of Gaussians.

We referred to a single Gaussian as $p(\xv)$.  We will now also call it $\mathcal{N}(\xv|\muv,\Sigmav)$.
$$
\mathcal{N}(\xv|\muv,\Sigmav) = p(\xv) = \frac{1}{2\pi^{d/2} |\Sigmav |^{1/2}}
      e^{-\frac{1}{2} (\xv-\muv)^T \Sigmav^{-1} (\xv - \muv)}
$$
where mean $\muv$ is a $D$-dimensional column vector and covariance
matrix $\Sigmav$ is a $D\times D$ symmetric matrix.

We define a mixture of these by summing up the values of a number of Gaussians, each one weighted by scalar value between 0 and 1.
$$
p(\xv|\muv,\Sigmav) = \sum_{k=1}^K \pi_k \mathcal{N}(\xv|\muv_k,\Sigmav_k),
$$
where $K$ is the number of Gaussian components in our mixture and $\pi_k$ are the mixing coefficients for which $0 \le \pi_k \le 1$ and \sum_{k=1}^K \pi_k = 1$.

Given $N$, $D$-dimensional data samples in $\Xv$, which would be $N\times D$, how do we find good values of $\piv$, $\muv$, and $\Sigmav$?

A common approach is to maximize the likelihood that the model would generate the data samples. What is this likelihood?  We saw it above when working with a single Gaussian.

$$
L(\Xv | \piv, \muv, \Sigmav) = \prod_{n=1}^N p(\xv_n | \muv, \Sigmav)
$$

To maximize this, we have to find where its derivative is equal to zero.  How do you take the derivative of a product?  Right, we don't.  Instead we will work with the derivative of the natural logarithm.
$$
\begin{align}
LL(\Xv | \piv, \muv, \Sigmav) & = \ln \prod_{n=1}^N p(\xv_n | \muv, \Sigmav)\\
& = \sum_{n=1}^N \ln p(\xv_n | \muv, \Sigmav)\\
& = \sum_{n=1}^N \ln \left ( \sum_{k=1}^K \pi_k \mathcal{N}(\xv_n|\muv_k,\Sigmav_k) \right )
\end{align}
$$




How would you calculate the derivative (gradient) of $LL$ with respect to its parameters?


$$
\begin{align}
\nabla_{\muv_k} LL(\Xv | \piv, \muv, \Sigmav) & = \sum_{n=1}^N \nabla_{\muv_k} \ln p(\xv_n | \piv, \muv, \Sigmav)\\
& = \sum_{n=1}^N \frac{1}{p(\xv_n | \piv, \muv, \Sigmav)} \nabla_{\muv_k} p(\xv_n | \piv, \muv, \Sigmav)\\
& = \sum_{n=1}^N  \frac{1}{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) } \nabla_{\muv_k} \sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) \\
& = \sum_{n=1}^N  \frac{1}{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) } \left ( \pi_k \nabla_{\muv_k} \sum_{j=1}^K \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) \right )\\
& = \sum_{n=1}^N  \frac{1}{\sum_{j=1}^K \pi_k \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }  \pi_k \nabla_{\muv_k} \sum_{j=1}^K \left ( \frac{1}{2\pi^{D/2} |\Sigmav_j |^{1/2}} e^{-\frac{1}{2} (\xv_n-\muv_j)^T \Sigmav_j^{-1} (\xv_n - \muv_j)} \right ) \\
& = \sum_{n=1}^N  \frac{1}{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }  \pi_k  \frac{1}{2\pi^{D/2} |\Sigmav_k |^{1/2}} e^{-\frac{1}{2} (\xv_n-\muv_k)^T \Sigmav_k^{-1} (\xv_n - \muv_k)} (-1)  \frac{1}{2} 2\;  \Sigmav_k^{-1} (\xv_n - \muv_k) (-1)\\
& = \sum_{n=1}^N  \frac{1}{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }  \pi_k  \frac{1}{2\pi^{D/2} |\Sigmav_k |^{1/2}} e^{-\frac{1}{2} (\xv_n-\muv_k)^T \Sigmav_k^{-1} (\xv_n - \muv_k)}   \Sigmav_k^{-1} (\xv_n - \muv_k) \\
& = \sum_{n=1}^N  \frac{1}{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }  \pi_k  \mathcal{N}(\xv|\muv_k, \Sigma_k)  \Sigmav_k^{-1} (\xv_n - \muv_k) \\
& = \sum_{n=1}^N  \frac{\pi_k  \mathcal{N}(\xv_n|\muv_k, \Sigma_k) }{\sum_{j=1}^K \pi_j \mathcal{N}(\xv|\muv_j,\Sigmav_j) }   \Sigmav_k^{-1} (\xv_n - \muv_k) \\
\end{align}
$$

As before, we can set this equal to zero and solve for each $\muv_k$.  
$$
\begin{align}
0 &= \sum_{n=1}^N  \frac{\pi_k  \mathcal{N}(\xv_n|\muv_k, \Sigma_k) }{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n\muv_j,\Sigmav_j) }   \Sigmav_k^{-1} (\xv_n - \muv_k) \\
&= \Sigmav_k^{-1}  \sum_{n=1}^N  \frac{\pi_k  \mathcal{N}(\xv_n|\muv_k, \Sigma_k) }{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }   (\xv_n - \muv_k) \\
&= \Sigmav_k \Sigmav_k^{-1}  \sum_{n=1}^N  \frac{\pi_k  \mathcal{N}(\xv_n|\muv_k, \Sigma_k) }{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }   (\xv_n - \muv_k) \\
&=  \sum_{n=1}^N  \frac{\pi_k  \mathcal{N}(\xv_n|\muv_k, \Sigma_k) }{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }   (\xv_n - \muv_k) \\
\end{align}
$$

We have just arrived at equation (9.16) of our textbook.

Following the text, the fraction in the previous equation can be labeled $\gamma_{nk}$ to represent the ratio of the weighted $k^{th}$ Gaussian to the sum of all of the Gaussians, for the $n^{th}$ sample.  When $\xv_n$ is very close to $\muv_k$, $\gamma_{nk}$ is close to 1.  (Actually, the text uses the notation $\gamma(z_{nk})$.)

Letting
$$
\begin{align}
\gamma_{nk} &=  \frac{\pi_k  \mathcal{N}(\xv_n|\muv_k, \Sigma_k) }{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) } 
\end{align}
$$
we get
$$
\begin{align}
0 &=  \sum_{n=1}^N  \frac{\pi_k  \mathcal{N}(\xv_n|\muv_k, \Sigma_k) }{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }   (\xv_n - \muv_k) \\
&=  \sum_{n=1}^N  \gamma_{nk}  (\xv_n - \muv_k) \\
\end{align}
$$
And from this, it is clear that
$$
\begin{align}
0 &=  \sum_{n=1}^N  \gamma_{nk}  (\xv_n - \muv_k) \\
&= \sum_{n=1}^N  \gamma_{nk}  \xv_n - \sum_{n=1}^N  \gamma_{nk} \muv_k \\
\sum_{n=1}^N  \gamma_{nk} \muv_k &= \sum_{n=1}^N  \gamma_{nk}  \xv_n \\
\muv_k  \sum_{n=1}^N  \gamma_{nk} &= \sum_{n=1}^N  \gamma_{nk}  \xv_n \\
\muv_k  N_k &= \sum_{n=1}^N  \gamma_{nk}  \xv_n \\
\muv_k  &= \frac{1}{N_k}\sum_{n=1}^N  \gamma_{nk}  \xv_n \\
\end{align}
$$
where $\gamma_{nk} =  \frac{\pi_k  \mathcal{N}(\xv_n|\muv_k, \Sigma_k) }{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }$ and $N_k = \sum_{n=1}^N  \gamma_{nk}$.

Doing a similar derivation for $\nabla_{\Sigmav_k}  LL(\Xv | \piv, \muv, \Sigmav)$ and setting it to zero allows us to derive

$$
\begin{align}
\Sigma_k &= \frac{1}{N_k} \sum_{n=1}^N \gamma_{nk} (\xv_n - \mu_k) (\xv_n - \mu_k)^T
\end{align}
$$

One more step...we still have to figure out what the $\piv_k$ values should be.  Again, take derivatives and set to zero.

$$
\begin{align}
\nabla_{\pi_k} LL(\Xv | \piv, \muv, \Sigmav) & =  \sum_{n=1}^N  \frac{1}{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) } \nabla_{\pi_k} \sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) \\
& =  \sum_{n=1}^N  \frac{1}{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) } \mathcal{N}(\xv_n | \muv_k,\Sigmav_k) \\
& = \sum_{n=1}^N  \frac{\mathcal{N}(\xv_n|\muv_k, \Sigma_k) }{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }  
\end{align}
$$

However, no where have we considered the constraint on $\pi$ values that they sum to 1.  We can take care of this using Lagrange multipliers, which are nicely explained at [this video tutorial](https://www.youtube.com/watch?v=ry9cgNx1QV8).

$$
\begin{align}
\nabla_{\pi_k} \left ( LL(\Xv | \piv, \muv, \Sigmav) + \lambda ( \sum_{j=1}^K \pi_j - 1) \right ) & =  \sum_{n=1}^N  \frac{\mathcal{N}(\xv_n|\muv_k, \Sigma_k) }{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }  + \lambda
\end{align}
$$
Setting this equal to zero, then multiplying both sides by $\pi_k$, results in
$$
\begin{align}
0 & =  \sum_{n=1}^N  \frac{\pi_k \mathcal{N}(\xv_n|\muv_k, \Sigma_k) }{\sum_{j=1}^K \pi_j \mathcal{N}(\xv_n|\muv_j,\Sigmav_j) }  + \pi_k \lambda\\ 
& =  \sum_{n=1}^N \gamma_{nk} + \pi_k \lambda\\
& =  N_k + \pi_k \lambda\\
-N_k &= \pi_k \lambda
\end{align}
$$
 

Summing over $k$, we get

$$
\begin{align}
\sum_{k=1}^K -N_k &= \lambda \sum_{k=1}^K \pi_k \\
-\frac{\sum_{k=1}^K N_k}{\sum_{k=1}^K \pi_k} &= \lambda
\end{align}
$$
Since $\sum_{k=1}^K \pi_k = 1$, 
$$
\begin{align}
-\sum_{k=1}^K N_k &= \lambda
\end{align}
$$

Plugging this solution for $\lambda$ into a previous equation, we have$$
\begin{align}
0 & =  \sum_{n=1}^N \gamma_{nk} + \pi_k \lambda\\
0 & =  \sum_{n=1}^N \gamma_{nk} - \pi_k \sum_{k=1}^K N_k\\
\pi_k &=  \frac{\sum_{n=1}^N \gamma_{nk}}{\sum_{k=1}^K N_k}\\
\pi_k &=  \frac{N_k}{\sum_{k=1}^K N_k}\\
\end{align}
$$
 




We now have three equations for $\muv_k$, $\Sigmav_k$, and $\pi_k$.  Unfortunately, they all depend on $\gamma_{nk}$.  

To deal with this problem, we will use the [Expectation-Maximization (EM) algorithm](http://campar.in.tum.de/twiki/pub/Main/MoritzBlume/EMGaussianMix.pdf) (also see the [Wikipedia entry](http://en.wikipedia.org/wiki/Expectation%E2%80%93maximization_algorithm)):
  * Pick initial values for $\muv_k$, $\Sigmav_k$ and $\pi_k$.
  * *Expectation Step:* Calculate $\gamma_{nk}$ using these values.
  * *Maximization Step:* Update $\muv_k$, $\Sigmav_k$ and $\pi_k$ using the results of the $\gamma_{nk}$ calculation.
  * Repeat last two steps until changes in the log likelihood function falls below some threshold.

And, now, da-da-da-daaaa, time for python implementations of each calculation.

\end{align}
$$

We need a way to iterate through pairs of the elements of $\muv$ and $\Sigmav$, together, one pair for each gaussian.

```python
def normald(X, mu, sigma):
    """ normald:
       X contains samples, one per row, N x D. 
       mu is mean vector, D x 1.
       sigma is covariance matrix, D x D.  """
    D = X.shape[1]
    detSigma = sigma if D == 1 else np.linalg.det(sigma)
    if detSigma == 0:
        print('mu is\n',mu)
        print('Sigma is\n',sigma)
        raise np.linalg.LinAlgError('normald(): Singular covariance matrix')
    sigmaI = 1.0/sigma if D == 1 else np.linalg.inv(sigma)
    normConstant = 1.0 / np.sqrt((2*np.pi)**D * detSigma)
    diffv = X - mu.T # change column vector mu to be row vector
    return normConstant * np.exp(-0.5 * np.sum(np.dot(diffv, sigmaI) * diffv, axis=1))[:,np.newaxis]
```

```python
# import pdb
def mixtureOfGaussians(X,K,nIterations=10,verbose=True):
    # X is N x D matrix of data samples, each being D-dimensional
    # Mu is K x D matrix of K mean vectors, each one D-dimensional
    # Sigma is K x D x D array of K covariance matrices, each one D x D
    # Pi is K x 1 vector of weightings for each Gaussian
    # Gamma is K x N array of responsibilities of Gaussian k for sample n
    
    N,D = X.shape
    randomSampleIndices = np.random.choice(range(N),K,replace=False)
    Mu = X[randomSampleIndices,:]
    print(randomSampleIndices, X.shape)
    print(Mu)
    Sigma = np.tile(np.eye(D),K).T.reshape((K,D,D))
    Pi = np.tile(1/K, K).reshape((-1,1))
    
    for i in range(nIterations):
        # Expectation Step
        gaussians = np.array([normald(X,mu.reshape((-1,1)),sigma) for mu,sigma in zip(Mu,Sigma)])
        gaussians = gaussians.squeeze()
        Gamma = ((Pi *gaussians) / np.sum(Pi * gaussians, axis=0))
        Gamma[np.isnan(Gamma)] = 0 # to take care of entries where above denominator is zero
        # pdb.set_trace()
        # Maximization Step
        Nk = np.sum(Gamma,axis=1).reshape((-1,1))
        Mu = np.dot(Gamma,X) / Nk
        diffsk = X[:,np.newaxis,:] - Mu
        for k in range(K):
            Sigma[k,:,:] = np.dot(Gamma[k,:] * diffsk[:,k,:].T, diffsk[:,k,:]) / Nk[k]
        Pi = Nk / np.sum(Nk)
        Pi = Pi.reshape((-1,1))
        if verbose:
            for k in range(K):
                print('Gaussian',k)
                print('Mu\n',Mu[k,:])
                print('Sigma\n',Sigma[k,:,:])
                print('Pi\n',Pi[k])
            
    return Mu, Sigma, Pi, Gamma
```

```python
m,s,p,g = mixtureOfGaussians(X,2,nIterations=3)
```

Let's track convergence by calculating $LL$:
$$
\begin{align}
LL(\Xv | \piv, \muv, \Sigmav) & = \ln \prod_{n=1}^N p(\xv_n | \muv, \Sigmav)\\
& = \sum_{n=1}^N \ln p(\xv_n | \muv, \Sigmav)\\
& = \sum_{n=1}^N \ln \left ( \sum_{k=1}^K \pi_k \mathcal{N}(\xv_n|\muv_k,\Sigmav_k) \right )
\end{align}
$$

```python
def mixtureOfGaussians(X,K,nIterations=10,verbose=True):
    # X is N x D matrix of data samples, each being D-dimensional
    # Mu is K x D matrix of K mean vectors, each one D-dimensional
    # Sigma is K x D x D array of K covariance matrices, each one D x D
    # Pi is K x 1 vector of weightings for each Gaussian
    # Gamma is K x N array of responsibilities of Gaussian k for sample n
    
    N,D = X.shape
    randomSampleIndices = np.random.choice(range(N),K,replace=False)
    Mu = X[randomSampleIndices,:]
    Sigma = np.tile(np.eye(D),K).T.reshape((K,D,D))
    Pi = np.tile(1/K, K).reshape((-1,1))
    
    LL = []
    for i in range(nIterations):
        # Expectation Step
        gaussians = np.array([normald(X,mu.reshape((-1,1)),sigma) for mu,sigma in zip(Mu,Sigma)])
        gaussians = gaussians.squeeze()
        Gamma = (Pi *gaussians) / np.sum(Pi * gaussians, axis=0)
        Gamma[np.isnan(Gamma)] = 0 # to take care of entries where above denominator is zero
        # Maximization Step
        Nk = np.sum(Gamma,axis=1).reshape((-1,1))
        Mu = np.dot(Gamma,X) / Nk
        diffsk = X[:,np.newaxis,:] - Mu
        for k in range(K):
            Sigma[k,:,:] = np.dot(Gamma[k,:] * diffsk[:,k,:].T, diffsk[:,k,:]) / Nk[k]
        Pi = Nk / np.sum(Nk)
        Pi = Pi.reshape((-1,1))
        LLi = np.sum(np.log(np.sum(Pi * gaussians, axis=0)))
        LL.append(LLi)
        if verbose:
            for k in range(K):
                print('Gaussian',k)
                print('Mu\n',Mu[k,:])
                print('Sigma\n',Sigma[k,:,:])
                print('Pi\n',Pi[k])
            
    return Mu, Sigma, Pi, Gamma, np.array(LL)
```

```python
m,s,p,g,ll = mixtureOfGaussians(X,2,nIterations=20,verbose=False)
plt.scatter(X[:,0],X[:,1], alpha= 0.5, s=50, c=np.argmax(g,axis=0));   
```
