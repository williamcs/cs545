The non-EEG channels are

   * EXG2	33	EOG vertical right

Typically, the EEG channels (indices 0 through 31) are referenced to the mean of the earlobes. That’s easy.  Let's also subtract the mean of each channel.

```python
ref = np.mean(eegb[36:38,:],axis=0).reshape((1,-1))
eegbr = eegb[:32,:] - ref
eegbr = eegbr.T
eegbr = eegbr - np.mean(eegbr,axis=0)
eegb.shape
```

They don't very much!  Let's use 800 samples for the trial length.  Since the sampling rate is 1000 Hz, 800 samples corresponds to 800 milliseconds.

Let's segment the data into trials.  Construct a list of lists being 800 samples from all channels plus 200 samples (200 ms) prior to the stimulus, to be used in minute.

```python
segments = []
labels = []  # 1 for target, 2 for nontarget letters
for start in starts:
    segments.append( eegbr[start-200:start+800,:] )
    labels.append( 1 if stimulus[start+1] == 16 else 2 )
segments = np.array(segments)
labels = np.array(labels)
segments.shape, labels.shape
```

Let's just look at the segments for the P4 channel.

```python
letterb['channels'].index('P4')
```

```python
p4 = 18
plt.figure(figsize=(10,10))
for i in range(80):
    plt.subplot(9,9,i+1)
    baselineCorrected = segments[i,200:,p4] - np.mean(segments[i,:200,p4])
    plt.plot(baselineCorrected,c=('red','blue')[labels[i]-1])
    plt.ylim(-50,50)
    plt.xticks([]) # to remove xtick labels
print('Red ones are target trials that should contain P300s.')
```

Let's add the other two target letters to this data.

```python
def segmentize(eegdict):   
    eeg = eegdict['eeg']['trial 1']
    eeg = np.array(np.array(eeg))
    ref = np.mean(eeg[36:38,:],axis=0).reshape((1,-1))
    eegr = (eeg[:32,:] - ref).T
    eegr = eegr - np.mean(eegr,axis=0)
    
    stimulus = eeg[40,:]
    stimulus = stimulus - np.min(stimulus)
    starts = np.where( np.logical_and(np.diff(stimulus) > 0, stimulus[1:] < 100 ))[0]
    segments = []
    labels = []  # 1 for target, 2 for nontarget letters
    for start in starts:
        segments.append( eegr[start-200:start+800,:] )
        labels.append( 1 if stimulus[start+1] == 16 else 2 )
    segments = np.array(segments)
    labels = np.array(labels)
    
    return segments,labels
```

```python
protocols = [d['protocol'] for d in data]
print(protocols)
protocols.index('letter-p')
```

```python
segmentsb,labelsb = segmentize(data[protocols.index('letter-b')])
segmentsp,labelsp = segmentize(data[protocols.index('letter-p')])
segmentsd,labelsd = segmentize(data[protocols.index('letter-d')])
segmentsb.shape,labelsb.shape, segmentsp.shape,labelsp.shape, segmentsd.shape,labelsd.shape
```

```python
segments = np.vstack((segmentsb,segmentsp,segmentsd))
labels = np.hstack((labelsb,labelsp,labelsd))
segments.shape, labels.shape
```

```python
p4 = 18
plt.figure(figsize=(20,20))
coefficients = []
for i in range(240):
    plt.subplot(16,16,i+1)
    baseline = np.mean(segments[i,:200,p4])
    baselineCorrected = segments[i,200:,p4] - baseline
    plt.plot(baselineCorrected,c=('red','blue')[labels[i]-1])
    plt.ylim(-50,50)
    # fit with polynomial
    xs = np.arange(len(baselineCorrected))
    X = np.vstack((np.ones(len(xs)), xs, xs**2, xs**3, xs**4, xs**5)).T
    w = np.linalg.lstsq(np.dot(X.T,X), np.dot(X.T,baselineCorrected.reshape((-1,1))))[0]
    eegfit = np.dot(X,w)
    coefficients.append(w)
    plt.plot(eegfit,'g-',lw=3,alpha=0.8)
    plt.xticks([]) # to remove xtick labels
print('Red ones are target trials that should contain P300s.')
```

```python
coefficients = np.array(coefficients).squeeze()
coefficients.shape
```

Let's just try to classify it with a neural network.

```python
import mlutils as ml
import neuralnetworks3 as nn
```

```python
segments[:,:,p4].shape
```

```python
Xtrain,Ttrain,Xtest,Ttest = ml.partition(segments[:,:,p4],labels.reshape((-1,1)),80,classification=True)
```

```python
Xtrain.shape,Ttrain.shape,Xtest.shape,Ttest.shape
```

```python
np.sum(Ttrain==1), np.sum(Ttrain==2), np.sum(Ttest==1), np.sum(Ttest==2)
```

```python
36/180, 24/72
```

```python
nnet = nn.NeuralNetworkClassifier(Xtrain.shape[1],50,2)
```

```python
nnet.train(Xtrain,Ttrain,nIterations=100)
```

```python
plt.plot(nnet.getErrorTrace());
```

```python
ml.printConfusionMatrix( ml.confusionMatrix(Ttest, nnet.use(Xtest), (1,2)), (1,2))
```

```python
nnet = nn.NeuralNetworkClassifier(Xtrain.shape[1],(50,50,5,5),2)
nnet.train(Xtrain,Ttrain,nIterations=200)
plt.plot(nnet.getErrorTrace());
c=ml.confusionMatrix(Ttest, nnet.use(Xtest), (1,2))
ml.printConfusionMatrix( c, (1,2))
(c[0,0]*c[0,3] + c[1,1]*c[1,3]) / (c[0,3]+c[1,3])
```
